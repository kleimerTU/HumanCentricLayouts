# Continue Training
Continue training a partially trained network using **--continue_train** and specifiyng the models to load using **--load_encoder** and **--load_decoder**. Epochs and learning rate are not saved with the model, so they need to be set manually.

```
!python train.py --epochs 500 --data_path ../0-data/4-pydata/input/bedroom --gpu 0 --save_network_every 5 --batch_size 1000 --no_plot --lr 0.001 --continue_train --load_encoder ./models/pretrained_encoder.pkl --load_decoder ./models/pretrained_decoder.pkl
```

# Other Datasets and Parameters
When training other datasets, the **--box_code_size** parameter needs to be set according to the first dimension of boxes.mat. The authors use a **--batch_size** of 10% of the dataset size. The parameters **--lr_decay_by** and **--lr_decay_every** can be used to adjust the learning rate during training.

```
!python train.py --epochs 500 --data_path ../0-data/4-pydata/input/living --gpu 0 --save_network_every 5 --batch_size 548 --no_plot --lr 0.001 --lr_decay_by 0.95 --lr_decay_every 20 --box_code_size 19
```
